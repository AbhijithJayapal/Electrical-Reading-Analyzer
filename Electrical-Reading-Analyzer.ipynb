```python
def classify_voltages(voltage):
    if voltage > 250:
        return "Overvoltage"
    elif voltage < 200:
        return "Undervoltage"
    else:
        return "Normal"



results = []

print("Enter voltage readings one by one. Type 'done' when finished.")

while True:
    user_input = input("Enter voltage: ")
    
    if user_input.lower() == "done":
        break
    
    try:
        voltage = int(user_input)
        status = classify_voltages(voltage)
        results.append(f"{voltage}V - {status}")
    except ValueError:
        print("Please enter a valid number (or 'done' to finish)")

for r in results:
    print(r)
```
